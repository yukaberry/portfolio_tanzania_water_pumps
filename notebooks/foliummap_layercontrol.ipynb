# Maps without layercontrol bar

```python
tanzania_map = folium.Map(location=[ -6.802353, 39.279556], zoom_start=6)
# instantiate a mark cluster object for the incidents in the dataframe
df_map= plugins.MarkerCluster().add_to(tanzania_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df20.latitude,df20.longitude, df20.status_group):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(df_map)

# display map
tanzania_map 
```

# Maps with LayerControl bar 

```python
##Step1##
# Create a map with "OpenStreet" tile
tanzania_map = folium.Map(location=[ -6.802353, 39.279556], zoom_start=6)


##Step2##
# Create objects for each label using FeatureGroup() and define Tick bar name
feature_group_0 = FeatureGroup(name="Function")
feature_group_1 = FeatureGroup(name="needs repair")
feature_group_2 = FeatureGroup(name="Non Function")

##step3##
# Create "function"'s markers 
# loop through the dataframe and add each data point to MarkerCluster
# ensure to use right DataFrame. This is for "Function" so used "df0_20".(20 rows of label "0" ="Funciton")
markers_0= plugins.MarkerCluster().add_to(feature_group_0)
for lat, lng, label, in zip(df0_20.latitude,df0_20.longitude, df0_20.status_group):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(markers_0)

# Create "Needs repair"'s markers 
markers_1= plugins.MarkerCluster().add_to(feature_group_1)
for lat, lng, label, in zip(df1_20.latitude,df1_20.longitude, df1_20.status_group):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(markers_1)

# Create "Non Function"'s markers 
markers_2= plugins.MarkerCluster().add_to(feature_group_2)
for lat, lng, label, in zip(df2_20.latitude,df2_20.longitude, df2_20.status_group):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(markers_2)

    
    
##Step4##
# Added objects to a map
feature_group_0.add_to(tanzania_map)
feature_group_1.add_to(tanzania_map)
feature_group_2.add_to(tanzania_map)

##step5##
# create " layerControl" object to be able to control tick/untick
LayerControl().add_to(tanzania_map)

# display map
tanzania_map 
```
